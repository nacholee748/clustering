```python
import numpy as np
n_clusters = 5
dt = np.random.randint(0,100, (100,2))
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt

# Initialize the clusterer with n_clusters value and a random generator
# seed of 10 for reproducibility.
clusterer = KMeans(n_clusters=n_clusters)
cluster_labels = clusterer.fit_predict(dt)

plt.subplot(121)
plt.title("Unclustered Data")
plt.scatter(dt[:, 0], dt[:, 1], s=15)
plt.xticks([])
plt.yticks([])
# plt.show()

plt.subplot(122)
# Based on predicted_labels, we assign each data point distinct colour
colors = ["#4EACC5", "#FF9C34", "#4E9A06", "m"]
for k, col in enumerate(colors):
    cluster_data = cluster_labels == k
    plt.scatter(dt[cluster_data, 0], dt[cluster_data, 1], s=15)

   
plt.title("Clustered Data")
plt.xticks([])
plt.yticks([])
plt.show()
```
